```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a smooth, less periodic function of two variables
def f(x, y):
    return 0.3 * x**3 - 0.4 * y**2 + 0.2 * x * y + 1

# Define the domain
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Pick a fixed y value to slice across x
y_slice = 0.3
z_slice = f(x, y_slice)

# Pick a random x value away from the boundary (inner 80% of the interval)
x_rand = np.random.uniform(-0.8, 0.8)
z_rand = f(x_rand, y_slice)

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface with more transparency
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5, edgecolor='none')

# Plot slicing red line (bold)
ax.plot(x, np.full_like(x, y_slice), z_slice, color='red', linewidth=4, label=f'ICE plot across feature x1')

# Plot bold black dot at (x_rand, y_slice, 0), well inside the plane
ax.scatter(x_rand, y_slice, 0, color='black', s=50, label='Analyzed observation')

# Vertical red dashed line connecting dot to curve
ax.plot([x_rand, x_rand], [y_slice, y_slice], [0, z_rand], 'r--', linewidth=2)

# Adjust point of view (clockwise rotation by 60 degrees)
ax.view_init(elev=30, azim=105)

# Labels and legend
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('F(x1, x2)')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function (our model)
def f(x1, x2):
    return 0.3 * x1**3 - 0.4 * x2**2 + 0.2 * x1 * x2 + 1

# Create synthetic dataset (10 observations)
n_obs = 10
x2_vals = np.random.uniform(-1, 1, size=n_obs)
x1_range = np.linspace(-1, 1, 100)

# Calculate ICE curves
ice_curves = []
for x2 in x2_vals:
    ice = f(x1_range, x2)  # vary x1, hold x2 fixed
    ice_curves.append(ice)

ice_curves = np.array(ice_curves)
pdp = np.mean(ice_curves, axis=0)

# Plot ICE curves
plt.figure(figsize=(10, 6))
for i in range(n_obs):
    plt.plot(x1_range, ice_curves[i], color='gray', alpha=0.6, label='ICE' if i == 0 else "")

# Plot PDP
plt.plot(x1_range, pdp, color='red', linewidth=3, label='PDP (average of ICEs)')

# Plot formatting
plt.xlabel('Feature x1')
plt.ylabel('Model Prediction f(x1, x2)')
plt.title('ICE and PDP Comparison for Feature x1')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a smoother, less periodic function
def f(x1, x2):
    return 0.3 * x1**3 - 0.4 * x2**2 + 0.2 * x1 * x2 + 1

# Define domain
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-1, 1, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Fixed point inside the plane
x1_fixed = 0.4
x2_fixed = -0.2
z_fixed = f(x1_fixed, x2_fixed)

fig = plt.figure(figsize=(14, 6))

# Variant 1: Red line across x1 (x2 fixed)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.5, edgecolor='none')
z_line1 = f(x1, x2_fixed)
ax1.plot(x1, np.full_like(x1, x2_fixed), z_line1, color='red', linewidth=4, label='ICE plot for feature x1')
ax1.scatter(x1_fixed, x2_fixed, 0, color='black', s=50, label='Analyzed observation')
ax1.plot([x1_fixed, x1_fixed], [x2_fixed, x2_fixed], [0, z_fixed], 'r--', linewidth=2)
ax1.view_init(elev=30, azim=105)
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('f(x1, x2)')
ax1.legend()

# Variant 2: Red line across x2 (x1 fixed)
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.5, edgecolor='none')
z_line2 = f(x1_fixed, x2)
ax2.plot(np.full_like(x2, x1_fixed), x2, z_line2, color='red', linewidth=4, label='ICE plot for feature x2')
ax2.scatter(x1_fixed, x2_fixed, 0, color='black', s=50, label='Analyzed observation')
ax2.plot([x1_fixed, x1_fixed], [x2_fixed, x2_fixed], [0, z_fixed], 'r--', linewidth=2)
ax2.view_init(elev=30, azim=105)
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_zlabel('f(x1, x2)')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the model function
def f(x1, x2):
    return 0.3 * x1**3 - 0.4 * x2**2 + 0.2 * x1 * x2 + 1

# Create synthetic dataset
n_obs = 10
x1_range = np.linspace(-1, 1, 200)
x2_vals = np.random.uniform(-1, 1, size=n_obs)

# Compute ICE and d-ICE curves
ice_curves = []
dice_curves = []

for x2 in x2_vals:
    ice = f(x1_range, x2)
    dice = np.gradient(ice, x1_range)  # Numerical derivative w.r.t x1
    ice_curves.append(ice)
    dice_curves.append(dice)

# Convert to array and compute d-PDP
dice_curves = np.array(dice_curves)
dice_pdp = np.mean(dice_curves, axis=0)

# Plot d-ICE and d-PDP
plt.figure(figsize=(10, 6))

# d-ICE curves
for i in range(n_obs):
    plt.plot(x1_range, dice_curves[i], color='red', alpha=0.5)

# d-PDP (average)
plt.plot(x1_range, dice_pdp, color='black', linewidth=3, label='d-PDP (avg. of d-ICEs)')

# Labels and formatting
plt.xlabel('Feature x1')
plt.ylabel('Derivative ∂f/∂x1')
plt.title('Derivative ICE (d-ICE) and d-PDP for Feature x1')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
