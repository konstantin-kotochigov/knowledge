Типичный action flow для treatment-моделей состоит из 2 шагов:

1. **Моделирование.** Моделируем результат выполнения некоторого действия. 

    Результат - это, разумеется, случайная величина (обзовём y). Она имеет свое распределение и может быть:
  - бинарной, например отклик/не отклик на рассылку
  - числовой, например выручка от предложенной скидки
  - категориальный, 
  
  В большинстве случаев, когда говорим о моделировании случайной величины, имеется в виду получение оценки для её матожидания. 

    $$E(y|X) \sim f(X)$$
    
    X в данном случае - это контекст выполнения действия: с какого типа клиентом оно выполняется, какие атрибуты самого действия, какие при этом используются внешние факторы и прочее.
 
  
2. **Оптимизация.** Определеяем свои действия так, чтобы максимизировать суммарный ожидаемый эффект от них. 

  - для формализации термина "суммарный эффект" выбрается мера качества (как правило, это бизнес-мера)
  - определяем набор действий X, которые нужно сделать и их параметры
  - при необходимости учитываем ограничения (например, нельзя коммуницировать с клиентом чаще, чем раз в месяц)

**Модель отклика**

Для кейса с моделью отклика. Задача: определеить кому показывать маркетинговое предложение.
1. моделируем вероятность отклика клиента X на предложение (веротяность отклика в данном случае есть матожидание бинарной случайной величины) 
2. Отбираем клиентов так, чтобы мы уложились в выделенный бюджет, и при этом матожидание числа откликнувшихся клиентов было максимальным.

Чуть более сложный вариант - при отборе клиентов стоит учитываем и стоиомсть коммуникации, и потенциальную выгоду.

Данные модели Ннкак не отвечают на вопрос, а как бы повели себя клиенты, если бы мы им ничего не рассылали. Возможно, они бы и так покупали.

В случае с рекламными предложениями проблема не так очевидна, но для кейса с оттоком она становится более понятной.

**Модель оттока**

По каждому клиенту определяем вертяность оттока. Задача уменьшить её для хороших клиентов, следовательно нужны какие-то скидки.

В машинном обучении мерой Lift принято называть показатель того, насколько наша модель работает лучше, чем некий baseline по умолчанию. При единичном лифте ценность модели тоже равна нулю, так как она не вносит никакой новой ценности.

### Uplift Modeling

Идея: Мы хотим моделировать сразу не ценность действия, а увеличение этой ценности по сравнению с бейзлайном (то есть lift). Под бейзлайном, как правило, понимается отсутствие действия.

В парадигме uplift оно называется Treatment.

Классическая картинка для модели отклика.

Для простоты будем считать, что у нас всего один товар. Отсортируем клиентов по убыванию веротяности отклика и условно разобъем все это множество на 2 сегмента (купят, не купят).

Добавим на эту картинку горизонтальную ось - а купят ли клиенты этот товар, если его НЕ рекламировать ("buy if not treated"). Тут каждый сегмент разбивается еще на 2.

Получившиеся 4 сегмента называются так:
- **Sleeping dogs** - хорошие клиенты, но очень не любят, когда им надоедают
- **Sure things**- купят при любом раскладе
- **Persuadables** - внушаемые
- **Lost causes** - не купят ни при каком раскладе

<img src="img/uplift3.png" width=250>

В такой поставновке становится очвеидно, что при выборе кому слать рекламу, нас интересуют далеко не все клиенты. Нам нельзя попасть на sleeping dogs - это потеря клиента - и нежелательно попадать на sure things и lost causes - это потеря затрат на коммуникацию.

В классической постановке мы таргатируемся ориентируясь только на $P(response|X)$:
<img src="img/uplift4.png" width=300>

В uplift постановке мы таргетируемся, ориентируясь на разницу $P(response|X,T) - P(response|NT)$.
<img src="img/uplift5.png" width=350>

Проиллюстрировать процесс uplift-таргетирования можно следующей картинкой:
<img src="img/uplift1.png" width=500>

В начале таргетируемся на лучший сегмент - Persuadables - они генерируют много инкрементальных продаж.
По мере расширения выборки подключаются сегменты sure things и lost_causes, никакого эффекта не приносят.
Затем подключаются sleeping dogs, которые дают негативный эффект.

На картинке виден ярко выраженный оптимум, это тот сегмент, который мы в итоге должны отобрать.
