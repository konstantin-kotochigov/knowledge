# Wilcoxon Signed Test
---


Wilcoxon signed test is a nonparametric (rank) test to check $H_0$ that there is no difference between 2 __paired__ groups

*Paired observations = for each instance there are 2 measurments (before / after)
<img src="img/paired.png" width=200>

To work with this kind of data we construct a single dataset of differences $X=X_A-X_B$

__Idea:__ irrespective of magnitudes of differences their directions (a > b or otherwise) must be distributed evenly - for each positive difference there must be correposnding negative difference. Therefore medians of positive and negative differences (sumed ranks) should coincide

<img src="img/wilcoxon_1.png" width=400>

If dataset A tends to exceed B, there gonna be more red lines and vice versa

### Algorithm
1. Observation = difference $X_A - X_B$
2. Merge data in a single pool, order it by absolute value
3. $T = T^+ - T^-$ where $T^-$ sum of ranks negative differences and $T^+$ positive differences

<img src="img/wilcoxon_signed_1.png" width=400>

Either of these three T statistics can be used to check p-values

# T statistic
There is no closed-form definition of the statistic distribution but:<br>it is continuous, symmetric and converges to normal

There are three ways to evaluate:
- exact test
- normal approximation
- permutation test


Why T converges to normal?
Because of CLT

Knowing $\mu$ and $\sigma$ we can compute z-score and get the p-values

Under $H_0$ there is no assymetry in rank distribution and statistic $T^-$ approximately equals $T^+$<br>Thus the expected value for $T^+$ is the median $\frac{1}{2} \big(\frac{1+n}{2}\big)$
