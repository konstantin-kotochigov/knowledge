#### 0.Введение
Как правило, существует много источников траффика. Кто-то переходит по ссылке из поиска, кто-то кликает на баннер, когда видит рекламу, кто-то приходит по прямой ссылке и так далее.

Источники объединяются в каналы (channel).

Разумеется, один пользователь может использовать посещать сайт много раз и пользоваться при этом разными каналами.

#### 1. Постановка

Цель атирибуции - оценить влияние каждого из каналов траффика на совершение конверсии.

Под конверсией мы понимаем либо покупку товара, либо любое другое целевого действия. В дальнейшем на основании полученной картины, мы можем делать выводы об эффективности каналов и о том, как можно перераспределить бюджет чтобы задействовать, например, больше эффективных каналов.

Процесс аналитики выглядит примерно следующим образом: 
- мы настраиваем кампанию
- устанавливаем нащи тэги на страницы сайта (ну или возможно только на landing page) 
- стартуем и проводим кампанию
- и по ее завершению собираем данные по ней и строим post-campaign аналитику

Разбиение на каналы индивидуально и зависит от постановки задачи. Это может быть как крупное разбиение (например, на уровне Programmatic, Social и Organic). Или это может быть что-то более детальное, с детализацией до конкретной площадки. Мы, как правило, мы считаем в двух вариантах - первый с детализацией до каналов, второй с детализацией до макроканалов. В любом случае список каналов определяется на этапе постановки задачи.

Что касается конверсий, они также могут выбираться исходя из текущих потребностей, часто это визит на сайт и покупка товара (тэг срабатывающий на странице Thank You Page после совершения покупки).

На выходе получаем некий отчет, в рамках которого мы выводим 6 групп показателей.



#### Цепочки каналов


Первое, что мы смотрим, это все цепочки, которые приводят к конверсии. И строим статистику по этим цепочкам. А именно, мы выводим топ наиболее часто встречающихся цепочек.

Цепочка - это последовательность касаний, а точнее каналов этих касаний, записанная в том порядке в котором они происходили. Касаниями могут быть показ рекламного банера или видео, клик, визит на сайт и любые другие действия, которые трекаются в рамках кампании.

При этом если идет несколько касаний подряд, то при отображении они объединяются в одну группу и рассматриваются как единый элемент цепочки. 

Такая визуализация помогает лучше понять, как именно выглядит путь к принятию решения о конверсии, и какое место в нем занимают конкретные каналы.


Цепочки могут быть
- одноканальными, если целиком состоят из одного канала
- многоканальными, если участвует несколько каналов. 

Как правило большая часть цепочек именно одноканальные.

Соотсевтсенно статистку мы выводим в разбивке по типу цеопчки, отдельно по одноканальным (как сейчас на картинке), отдельно многоканальным.

Для одноканальных здесь все просто, мы выводим долю каждого канала среди всех одноканальных цепочек.

Для многоканальных выводим сами цепочки (здесь приведен Top-10 цепочек) и их доля в общей массе цепочек.




#### Эффективная частота
Здесь мы определяем кол-во касаний от первого показа до конверсии и выводим распределние этого количества. То есть здесь например, в 64% цепочек достаточно одного касания для совершения визита. Внизу подписывается кол-во цепочек, на базе которызх посчитана статистика. 

Данная статистика выводится отдельно для каждого канала. Их можно наложить, их можно сравнить между собой, посмотреть сколько касаний достаточно сделать, чтобы в они приносили основную часть конверсий. И здесь же можно вывести рекомендуемое кол-во касаний по каждому из каналов. Считается как количество касаний, достаточных для 80 или 90% конверсий.



#### Время до конверсии
Далее мы считаем среднее время от первого контакта до конверсии. Здесь, я думаю, все прозрачно. Мы можем сравнить разные каналы, посмотреть, какие из них предполагают более быструю конверсию, а какие более долгую. Часто разница бывает довольно существенной.




#### Вклады каналов

Центральная цель анализа - определеить эффективность каналов, поэтому на данном шаге мы рассчитываем вклады каждого канала в совершение конверсии.

Что такое вклад? У нас есть цепочка касаний, есть конверсия. Задача - наиболее честным образом распределить ценность конверсии между этими каналами. 

Под ценностью конверсии чаще всего понимается одна из двух вещей:
- факт покупки (то есть мы распределяем просто единичную массу)
- либо стоимость покупки (то есть мы распределяем деньги которые данная транзакция принесла)

Соответственно вклад канала в конверсию - это и есть доля ценности конверсии, которая приодится на данный канал.

Если цепочка, как мы это называем, одноканальная (то есть например, клиент увидел рекламу на Youtube, кликнул, перешел на Landing Page, купил), то с определением вклада здесь все просто -  100% ценности относится к этому каналу, в данном случае к Youtube. 

Когда же каналов несколько (то есть мы имеем дело с многоканальными цепочками), все становится немножко сложнее - нужно понять, что и в какой мере повлияло на решение клиента о конверсии.

Существует несколько подходов как иименно распределять эту ценность и в рамках отчета мы рассчитываем все известные модели атрибуции. В отчет, как правило, идут 4 наиболее популярные: FT, LT, Linear и Markov. Далее расскажу чуть подробнее о каждой модели:

#### First Touch и Last Touch
Модель первого касания (First Touch)
В рамках данной модели весь вклад относится к тому каналу, с которого началась цепочка. То есть конверсия атрибуцируется к тому касанию, когда произошло первое взаимодействие с рекламируемым продуктом.
Модель последнего касания (Last Touch)
В рамках данной модели весь вклад относится к тому каналу, которым цепочка закончилась. То есть конверсия атрибуцируется к тому касанию, когда произошло последнее взаимодействие перед совершением покупки.

#### Модель Linear
В рамках данной модели весь вклад распределется межу всеми каналами, участвующими в цепочке пропорционально тому, сколько раз данный канал участвовал в цепочке. То есть при наличии цепочки, например {MyTarget, MyTarget, VK, VK, GPMD}, ценность конверсии будет распределена таким образом, на MyTarget и VK по 40% конверсии, на GPMD оставшиеся 20% конверсии

#### Модель Markov
Далее модель Маркова (Markov Model)

На практике, из-за того что каналы встречаются не в случайном порядке, а обычно зависят друг от друга, линейная модель не всегда может точно оценить распределение каналов. Поэтому часто используют модель Маркова.

Модель назвается Марковской из-за того, что цепочки по сути являются цепями Маркова - то есть вероятность увидеть канал в цепочке зависит только от того, какой канал шел сразу перед ним.

Данное свойство позволяет записать эти вероятности в виде матрицы переходов между каналами, а эта матрица в свою очередь может быть  визуализирована в виде простого направленного графа переходов, который собственно и показан на картинке

Здесь каждая вершина - это канал взаимодействия, а ребро описывает веротяность перехода в данном напрвлении (то есть от одного канала к другому). При этом обращаю внимание, что:
- добавляются 2 специальные вершины, это (начало цепочки) и (конверсия)
- есть вероятность остаться в той же врешине
- cумма вероятностей всех переходов равна 1.0

Видим например, что после касания через канал Segmento в 50% случаев следует заход через канал VK
А после просмотра на Youtube тоже в 50% случаев следует конверсия.

#### Модель Markov :: Removal Effect
Далее мы считаем, как повлияет на вер-ть конверсии удаление каждого из каналов. Для этого поочередно как бы выключаем каждый канал, пересчитываем вероятность конверсии (она считается просто математически по этой вероятностной сети), и выводим, насколько уменьшится веротяность конверсии. Этот показатель называется Removal Effect, он отдельно также выводистя у нас, и сам по себе он также оценивает “важность” канала.

Далее нормируем посчитанные removal эффекты по всем каналам и получаем распределение ценности конверсий по всем каналам.

В некотором смысле модель Маркова можно рассматривать как более универсальную модель, которая в сравнении с First Touch и Last Touch учитывает также взаимодействия между каналами.




#### Сравнительная эффективность
Далее также оцениваем каналы с точки зрения конверсионности. 

Вклады каналов по атрибуции важный показатель, но он не всегда дает полную картинку, поскольку высокий вклад может быть следствием как хорошей конвертируемости канала, так и просто большей доли канала в общем миксе. 

Чтобы учесть оба показателя, мы выводим следующий график. По оси X у нас посчитано отношение числа конверсий к числу показов. По оси Y посчитана доля данного канала среди всех показов.

Смысл этой визуализации - вывести на одной карте все каналы. Чтобы, во-первых, визуально было видно, как они соотносятся друг с другом по эффективности, и во-вторых чтобы можно было определить зоны роста.

Зоны роста - это в первую очередь те каналы, которые относительно мало используются, но при этом демонстрируют высокую конверсионность. Естественно, такие каналы разумно использовать больше, чем они используются сейчас.

И аналогично можно найти анти зоны роста - те каналы, которые на текущий момент съедают много бюджета, но при этом показывают более низкую эффективность по сравнению с остальными.
