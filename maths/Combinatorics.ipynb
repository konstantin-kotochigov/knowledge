Permuation with repeated elements.

Number of words of length n that can be created using a bag of letters $\{a:n_a,b:n_b \dots z:n_z\}$:
$$\frac{n!}{n_a!n_b! \dots n_z!}$$

#### Task 1
Number of ways to choose 4 employees from a 30 people (including 2 women and 28 men), given that both women are selected

$$C_{28}^{2} = 378$$

If we were asked the probability of selecting those 2 women

$$\frac{C_{28}^{2}}{C_{30}^{4}} \approx 1.5\%$$

#### Task 2
Number of ways to select 2 key roles from a comitee of 7 people

- Order matters
- No repetition

$$A(7,2) = \frac{7!}{(7-2)!} = 6 * 7 = 42$$

#### Task  3
Number of ways to assign 6 classes to 3 tutors (each must be assigned exactly 2)

Number of ways to assign 2 classes to first
$$C(6,2)$$

Number of ways to assign 2 classes to second:
$$C(4,2)$$

Third has no choice

So,
$$C(6,2) * C(4,2)$$

#### Task 4
There is an urn with 100 tickets. 5 successful.
Find number of ways to select 5 tickets, given that 3 were winning.

We know 3 tickets were winning, but which ones?
There are $C(5,3)$ options.

Each option comes along with 2 unsucsessgul tickets. But which ones?
There $C(95,2)$ options.

So, the answer is
$$C(5,3) * C(95,2)$$

#### Task 8

Find number of ways to choose 9 cakes if there are 3 kinds of them.

- order does not matter
- with repetition
- k > n

We got a formula when k > n

$$\hat{C}(3,9) = C(3+9-1,9) = \frac{11!}{2!9!}$$

#### Task 9

Find number of 4-digit numbers constructed from a set of 4 digits {0,3,4,5} given that each digit is used only once.

- order matters
- without repetition (each digit used once)
- k = n => it's called permutation

But we need to subtract numbers starting with 0 (they are impossible).

$$P(4) - P(3) = 24 - 6 = 18$$

3 ways to choose the first digit, 3! ways to distribute 3 other digits:

$$3 * (3!) = 18$$
