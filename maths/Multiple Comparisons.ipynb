In a simple scenario we make one comparison. In practice there are cases when it's important to make multiple comparisons.

When we conduct a statistical test, we set alpha to be equal to 0.05. It means that we expect False Alarm errors in 5% cases.

Familywise error rate - probability of observing <u>at least</u> one error



### Sidak correction
If we assume all tests are independent, we can calculate the exact value of the Familywise error rate:

$$\alpha_{FW} = 1 - (1-\alpha_{TW})^n$$

And vice versa - given the desired familywise error $\alpha_{FW}$, we can get the testwise error:

$$\alpha_{TW} = 1 - (1-\alpha_{FW})^{1/n}$$

The correction based on this expression is called Sidak correction.

### Bonferronni correction

In practice however we can use a simpler expression, which is called Bonferronni correction:

$$\alpha_{TW} = \frac{\alpha_{FW}}{n}$$

Because of its simplicity, it became way more popular in practice.

- Notice that Bonferonni alphas are simply the first order Taylor expansion of the Sidak alphas
- It's an upper bound, not exact value
- It can be derived also from Boole inequity

    Boole inequity = the upper bound of the Probability of union of events can be dervided as a sum of individual Probabilities.

    And it transforms to equity when all events are negatively correlated.
    
    In multiple comparison setting we similarly compare $m_0$ p-value tests:

$$\text{FWER} = P\left\{ \bigcup_{i=1}^{m_0}\left(p_i\leq\frac \alpha m \right) \right\} \leq\sum_{i=1}^{m_0}\left\{P\left(p_i\leq\frac \alpha m\right)\right\} =  m_0 \frac \alpha m \leq \alpha$$


- Bonferonni correction does not require independence => you can use it for correlated tests

### Test correlation

When tests in a family are correlated:
- it's **good** for us in terms of the error rate, since we don't need to correct alphas a lot. Familywise error will be close to Testwise error.
- It's **bad** in terms of confidence, if they make an error, they make it together

### Bonferonni vs Sidak
Bonferronni:
- is more conservative
- less stringent

Sidak:
- is more powerful => detects more discoveries
- requires independence




### Holm-Bonferonni correction (1979)

So far we used constant alphas that were dependent only on the number of tests.

Here the authors dynamically assign alphas according to their p-values ranking.

Method outline:
1. Sort hypotesis $H_1 H_2 \cdots H_m$ in ascending p-value order
2. Assign individual alpha according to the ranking of corresponding p-value
$$\alpha_{TW} = \frac{\alpha_{FW}}{n-(k-1)}$$
For example, if there are $m$ tests, this gives the following sequence of alphas:

$$\frac{\alpha}{m},  \frac{\alpha}{m-1}, \ldots ,  \frac{\alpha}{2},  \frac{\alpha}{1}$$
