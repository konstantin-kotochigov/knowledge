Tuple is a fixed set of stored data
```python 
t = (1,2,3)
```

Tuples in Python are:
1. __ordered__<br>access to elements by index only<br>```t[0], t[1]```<br><br>
2. __immutable__<br>have to be recreated to update - main purpose to store data, not manipulate<br><br>
3. __different types allowing__<br>```x=(1,'a',x)```<br><br>
4. __duplicate allowing__<br>tuples are meant for defining structure, not values<br>```x = (1, 1, 2, 2, 3, 3)```<br>

### Construction
- t = (a,b,c)<br>literal based construction - round braces<br><br>
- t = tuple((a,b,c))<br>explicit construction - for better readability<br><br>
- t = (a)<br>not a tuple<br><br>
- t = (a,)<br>one element tuple<br><br>


### Access
- range access - OK
- negative indexing - OK
- dot access ```t.0``` is not allowed

iteration is also OK

### Methods

Check existence
```python 
x in t
```

Count occurences
```python 
t.count(val)
```

Find occurences
```python
t.index(val)
```

### Concatenation
- Z = X + Y
- X += Y
- Y = X * 3<br>tuple repetition - Y = (X ... X ... X)

unpacking
```
(a,b,c) = t
```

partial unpacking
```
(a,*b) = t
```
b = list of remaining values
