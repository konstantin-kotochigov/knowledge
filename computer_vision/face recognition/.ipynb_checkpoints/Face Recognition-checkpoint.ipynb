Классификация на K классов (по чиcлу людей) - плохо, потому что всего по паре примеров каждого класса и нормально модель не обучится. Такой сеттинг называется one-shot learning.

Лучше делать - similarity/distance learning, обучать общую фуникцию близости.

Для этого используется сиамская сеть из двух CNN-ок:

В сиамской сети все веса расшарены сежду двумя ветками. Традиционно в сиамских сетях используется triplet loss-ы.

### Triplet Loss

Triplet Loss - обучение на тройках (anchor, positive, negative). В свое время метрика вводилась специально для задач Metric Learning (Distance Learning) - обучать корректное вычисление растояний.

Расстояние между похожими картинками должно быть меньше расстояния между разными картинками:
d(A,P) < d(A,N)

Чтобы сеть не выводила всегда нули, требуем чтобы еще был зазор
d(A,P) < a + d(A,N)

Сами расстояния - обычные квадратичные нормы
d(A,P) = L2 = ||f(xA) - f(xP)||2
d(A,N) = L2 = ||f(xA) - f(xN)||2

# Gabor Filters
Фильтр Габора - 2-мерный фильтр, который используют при описании изображений и который заточен на детектирование регулярных текстур.

Обычно используют сразу пачку таких фильтров.

Он довольно мудрено записывается. Это комбинация (1) гауссианы и (2) синусоиды.

Синусоида вообще записывается в виде комплексной экспоненты.
Параметры:
- скорость затухания гауссианы
- эллипсоидность науссианы
- частота синусооиды
- смещение
