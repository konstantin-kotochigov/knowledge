# T-SNE (2008)

T-SNE stands for T-stochastic neighbotshood embedding. It's a method for visualizing high-deimensional data introduced in 2008 by Hinton.

__Idea:__ find a projection onto latent space that preserves distances between neighboring points as much as possible

## Source space
---

For each point in a dataset let's calculate Gaussian similarity to all other points


<img src="img/tsne_gaussian.png" width=200>

The choice of Gaussian shows that we care only about the nearby neighbors

The spread hyperparameter $\sigma$ is to be adjusted further

<img src="img/tsne_gaussian_sigma.png" width=400>
After we built a matrix of similarities let's normalize them row-wise to form a proper probability distribution P for each point

$$P(i,j)=\frac{e^{-||x_i -x_j||^2 / 2\sigma_i}}{\sum_j e^{-||x_i -x_j||^2 / 2\sigma_i}}$$

Still our similarity is not symmetric ($p_{ij} \neq p_{ji}$) => let's construct a new measure as an average of two similarities:<br>
$p_{ij} = p_{ji} = \frac{1}{2} \cdot \big[p_{ij} + p_{ij}\big]$

Note that denominator in formula remains the same, so it's still a probability distribution.

## Latent space
---

In the latent space we'll use almost the same simlarity model with the only difference - we use Student t-distriubtion instead of Gaussian

<img src="img/tsne_student.png" width=200>

The choice of Student is explained by its heavier tails - this gives more flexibility to a model to rearrange dissimilar points. <br>T-distribution gives the first letter to the method name

<img src="img/tsne_similarities.png" width=450>

So, if we find a good projection, P and Q will be close to each other.


## Mapping model
---

We are trying to preserve "neighborhood distribution" => obvious choice for the loss function would be $D_{KL}$

$C = \sum_i D_{KL}(P_i||Q_i) = \sum_i P_i \, \cdot \log{\bigg(\frac{P_i}{Q_i}\bigg)}$


## Optimization
---
The optimization problem is solved iteratively using Gradient Descent
<img src="img/gradient_Step.png" width=400>


Some details:
- optimization is done with momentum $\alpha$
- typical process length = 1000 iterations
- higher learning rate $\mu$ at first 250 iterations to faster build strcture

## How to choose $\sigma$
---

The "spread" parameter $\sigma$ is chosen to correspond the desired **Perplexity** of the model

Recall that __Perpelxity__ of the distribution =  $2^H = 2^{-\sum p \, \log{p}}$ where H = entropy of the distribution (amount of randomness). Peaky distriubtions have lower entropy

If we choose P to be "narrow" distribution (low perplexity) we focus on local structure of the data. High perplexity = we focus on global structure

## Barnes-Hut approximation
---
Naive implementation has $O(n^2)$ complexity - we have to build an $(NxN)$ similarity matrix

Barnes-Hut uses a quad-tree to store the dataset => for faraway points we can approximate them as clusters of points. This significantly reduces the amount of computation and also makes it adjustable

<img src="img/quadtree.png" width=300>

## Pros & Cons
---

- On standard datasets (MNIST) t-SNE works significantly better than PCA
- Requires more computational resources
- Because of its complexity and existence of a hyperparameter, this method is way harder to tune. Each run can lead to different result

## Scikit-Learn Example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer

bunch = fetch_20newsgroups()
X = bunch['data']
y = bunch['target']

X = TfidfVectorizer(max_features=30000).fit_transform(X)
X_embedded = TSNE(n_components=2).fit_transform(X)

df = pd.DataFrame(X_embedded, columns=['x','y'])
df['c'] = y

df[0:10000].plot(kind='scatter', x='x', y='y', c='c', cmap=plt.get_cmap('Paired'), figsize=(15,10))
```
