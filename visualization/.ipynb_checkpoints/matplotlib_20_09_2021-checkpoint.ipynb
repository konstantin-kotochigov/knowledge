```python
import pandas as pd

x = numpy.random.normal(0,1,25)
y = numpy.random.normal(0,1,25)
c = numpy.random.randint(0,5,25)
s = 25*numpy.random.randint(1,5,25)

df = pd.DataFrame({"x":x, "y":y, "c":c, "s":s})

plt.scatter(x=df.x, y=df.y, c=df.c, s=df.s, edgecolor='black')

plt.grid(alpha=0.25)
plt.colorbar()

for row_index, row in df.iterrows():
  plt.text(x=row['x']+0.1, y=row['y']+0.05, s=str(row_index))
```

```python
from sklearn.datasets import load_boston

df_boston_all = load_boston()
```

```python
X = df_boston_all['data']
y = df_boston_all['target']
feature_names = df_boston_all['feature_names']

df = pd.DataFrame(X, columns=feature_names)
correlation_map = round(df.corr(), 2).values
```

```python
import matplotlib

fig,ax = plt.subplots()
plt.set_cmap(matplotlib.cm.get_cmap("coolwarm"))
ax.imshow(correlation_map, aspect='auto')

ax.set_xticks(range(len(feature_names)))
ax.set_yticks(range(len(feature_names)))

ax.set_xticklabels(feature_names, rotation=45, ha='right')
ax.set_yticklabels(feature_names)

for i in range(len(feature_names)):
  for j in range(len(feature_names)):
    ax.text(i,j, correlation_map[i,j], color='black', fontsize=9, ha='center', va='center')
```
