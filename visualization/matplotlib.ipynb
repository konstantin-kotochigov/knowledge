### Advanced chart types

Plots above comprise the most of data visualization. Though sometimes more sophisticated graphs might be a better choice

Radar plot - good for comparing structure of diffrerent objects
<img src="img/radar.jpeg" width=350>

- Tree plot  - describes complex hierachical structures
- Flow plot - describes strcuture (same as tree but allows cycles)

__Pie chart__ is a one-dimensional version of a bar-chart. Mostly used on KPI dashboards/scorecards and in business presentations when you need to describe structure or distribution. Quite useless since round shapes are harder to perceive. In most cases barchart is a better option.

<img src="img/pie.png" width=500>

__Normalized plot__ - describes dynamics of a structure. It is a combination of barchart and lineplot. Quite useless because it often fails to underline slight changes, which usually comprise the dynamics of most real world indicators.

Another popular way of describing data is presenting it as __tables__. They are simpler to design but may require more effort to perceive (depends on amount of data in it). Do not use tables if the conent cannot be read within seconds

The worst way of describing the results is __plain text__. 
```
As you can see, the sales have soared to the level of 156.000 which is 24.5% higher then previous period, but still lower than yearly average. Whereas the sales of Sacramento department remained on more or less the same level with 124500 per month. In the next period.
```
Why - because no one would ever read it and even fewer would remember its contents. Use text only for conclusions (unless you write a telegram)

### Drawing modes

Typical Matplotlib canvas consists of one or multiple charts.

There are two drawing modes:

- drawing on a canvas (commands are addressed via pyplot object)

        plt.scatter()


- drawing on a specific chart (commands are addressed via axes object)
    
        fig,ax = plt.subplot(1,1)
        ax.scatter()

First mode is more suitable for simple graphs, whether second is more useful when you design a multiplot chart with many plots on it.

Methods are almost the same, though here are some exceptions - it's better to always check the tutorial. For example, if we want to add axis ticks on the graph:
- plt.xticks() 
- ax.set_xticks()

Basic pyplot commands:
- __plot(y)__ draws a sequence of points connected with line<br>ommited X coordinates are assumed to be range(len(y))<br><br>
- __plot(x,y)__ draws a set of points (x,y) connected with line<br>points are connected in the order thay come => it does not have to be 1D plot or time series

When creating plots Matplotlib treats input data as numpy arrays (sequences of numbers). It also allows other types, since it always implicitly tries to convert input data to numpy array. 

More formally, X and Y have to be [array-like](https://stackoverflow.com/questions/40378427/numpy-formal-definition-of-array-like-objects). Array-like is any sequence of numbers that can be converted to ndarray - for example, Python list or pandas Series object

### Point annotations

````python
import pandas as pd
x = numpy.random.normal(0,1,25)
y = numpy.random.normal(0,1,25)
c = numpy.random.randint(0,5,25)
s = 50 * numpy.random.randint(0,5,25)
df = pd.DataFrame({"x":x, "y":y, "c":c, "s":s})

plt.scatter(
    x = df.x,
    y = df.y,
    c = df.c,
    s = df.s,
    edgecolor='black',
    alpha=0.5
)

plt.grid(True,alpha=0.5)
plt.colorbar()

for row_index, row in df.iterrows():
    plt.text(row['x'] + 0.1, row['y']+0.05, str(row_index))
````

### Visualization of Correlation matrix

````python
from sklearn.datasets import load_boston
import matplotlib

df_boston_all = load_boston()

X = df_boston_all['data']
y = df_boston_all['target']
feature_names = df_boston_all['feature_names']

df = pd.DataFrame(X, columns=feature_names)
correlation_map = round(df.corr(), 2).values

fig, ax = plt.subplots()
plt.set_cmap(matplotlib.cm.get_cmap("coolwarm"))
im = ax.imshow(correlation_map, aspect='auto')
ax.set_xticks(numpy.arange(len(feature_names)))
ax.set_yticks(numpy.arange(len(feature_names)))
ax.set_xticklabels(feature_names)
ax.set_yticklabels(feature_names)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

for i in range(len(feature_names)):
    for j in range(len(feature_names)):
        ax.text(i, j, correlation_map[i,j], color='w', fontsize=9, ha="center", va="center")
````

### Custom Grid Layout

````python
fig3 = plt.figure(constrained_layout=True)
gs = fig3.add_gridspec(3, 3)

f3_ax1 = fig3.add_subplot(gs[0, :])
f3_ax1.set_title('gs[0, :]')

f3_ax2 = fig3.add_subplot(gs[1, 0])
f3_ax2.set_title('gs[1,0]')

f3_ax3 = fig3.add_subplot(gs[1, 1:])
f3_ax3.set_title('gs[1,1:]')
````
