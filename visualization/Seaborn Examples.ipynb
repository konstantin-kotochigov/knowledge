Ноутбук в процессе разработки

```python
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_wine

sns.set_style("whitegrid")
```

```python
x = np.random.normal(size=100)

df = load_wine()
df = pd.DataFrame(df['data'], columns=df['feature_names'])
```

Построить гистограмму: distplot (distribution plot)

Параметры:
- data (numpy-array, Series, list)
- bins - гранулярность
- hist / kde / rug - что показывать на графике
- color
- norm_hist - если kde не строится, то нормировать или нет

```python
sns.distplot(x, bins=20, hist=True, kde=True, rug=True, norm_hist=False)
```

```python
sns.pairplot(df.iloc[:,0:5])
```

```python
sns.relplot(x="alcohol", y="malic_acid", data=df);
```

```python
sns.scatterplot(x="alcohol", y="malic_acid", data=df);
```

```python
sns.lineplot(x="alcohol", y="malic_acid", data=df);
```
